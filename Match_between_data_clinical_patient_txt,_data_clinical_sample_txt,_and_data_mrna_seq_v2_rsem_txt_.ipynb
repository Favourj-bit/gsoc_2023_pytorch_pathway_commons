```python
import pandas as pd
```

```python
data_sample = pd.read_csv("/content/drive/MyDrive/gsoc data/data_clinical_sample.txt", sep="\t")
```

```python
data_sample.head()
```

```python
ds1 = data_sample.drop([0,1,2,3])
ds1.head()
```

```python
ds1.shape
```

```python
# Extract the columns of interest
sample1 = ds1['Cancer Type Detailed']
sample2 = ds1['Cancer Type']

# Check if the two columns contain the same values
are_equal = sample1.equals(sample2)

if are_equal:
    print("The columns contain the same values.")
else:
    print("The columns do not contain the same values.")
```

```python
ds1.isnull().sum()
```

```python
cleaned_ds = ds1.dropna(axis=1, how='all')
cleaned_ds.isnull().sum()
```

```python
cleaned_ds = cleaned_ds.fillna(method='ffill')
```

```python
cleaned_ds = cleaned_ds.drop(columns='Cancer Type')
```

```python
cleaned_ds.head()
```

```python
data_patient = pd.read_csv("/content/drive/MyDrive/gsoc data/data_clinical_patient.txt", sep="\t")
```

```python
dcp = data_patient.drop([0,1,2,3])
dcp.head()
```

```python
dcp.isnull().sum()
```

```python
dcp['Overall Survival (Months)'] = dcp['Overall Survival (Months)'].fillna(method='ffill')
```

```python
dcp = dcp.dropna(how='any', axis=1)
```

```python
dcp.head()
```

```python
merged_df = pd.merge(cleaned_ds, dcp, on='#Patient Identifier')
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
# Assuming your DataFrame is named 'df'
columns_to_keep = ['#Patient Identifier', 'Sample Identifier', 'Overall Survival (Months)']
used_df = merged_df[columns_to_keep]
```

```python
used_df.head()
```

```python
used_df.shape
```

```python
df = pd.read_csv("/content/drive/MyDrive/gsoc data/data_mrna_seq_v2_rsem.txt", sep="\t")
```

```python
patients = df.copy()
patients.head()
```

```python
patients = patients.drop('Entrez_Gene_Id', axis=1)
```

```python
patients.shape
```

```python
patients.isnull().sum()
```

```python
patients = patients.dropna()
```

```python
identifiers = patients['Hugo_Symbol'].unique()
print(len(identifiers))
```

```python
# Check for duplicate 'Hugo_Symbol' values
duplicates = patients['Hugo_Symbol'].duplicated(keep='last')

# Drop rows with duplicate 'Hugo_Symbol' or same symbol as the previous row
new_df = patients[~duplicates]

# Verify the updated DataFrame shape
print(new_df.shape)
```

```python
new_df.shape
```

```python
new_df_transposed = new_df.set_index('Hugo_Symbol').transpose()
new_df_transposed.head()
```

```python
# Merge the datasets based on the patient identifiers
patient_df = pd.merge(used_df, new_df_transposed, left_on='Sample Identifier', right_index=True)
```

```python
patient_df.shape
```

```python
patient_df.head()
```

```python
patient_df = patient_df.drop('#Patient Identifier', axis=1)
```

```python
patient_df.set_index('Sample Identifier')
```

```python
cBio = pd.DataFrame(patient_df)
```

```python
cBio.to_csv('cBio.csv', index=False)
```
